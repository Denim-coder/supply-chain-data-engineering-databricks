```python
raw_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv("/Volumes/workspace/default/raw_food_supply/food_supply_chain.csv")
)

display(raw_df)
```

```python
from pyspark.sql.functions import col

bronze_df = raw_df.select(
    col("Product_ID").alias("product_id"),
    col("Product_Name").alias("product_name"),
    col("Supplier_Name").alias("supplier_name"),
    col("Origin_City").alias("origin_city"),
    col("Destination_Warehouse").alias("destination_warehouse"),
    col("Quantity_Shipped").alias("quantity_shipped"),
    col("Unit_Cost").alias("unit_cost"),
    col("Total_Cost").alias("total_cost"),
    col("Shipping_Mode").alias("shipping_mode"),
    col("Shipment_Date").alias("shipment_date"),
    col("Delivery_Date").alias("delivery_date"),
    col("Quality_Check_Status").alias("quality_check_status")
)

display(bronze_df)
```

```python
raw_df.write.format("delta") \
  .mode("overwrite") \
  .saveAsTable("scf_bronze.raw_supply.bronze_food_supply_chain")
```

```python
%sql
COMMENT ON TABLE workspace.default.bronze_food_supply_chain IS
'Bronze layer table containing raw food supply chain data ingested from a managed volume';
```
