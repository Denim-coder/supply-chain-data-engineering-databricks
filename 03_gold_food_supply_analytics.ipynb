```python
%sql
CREATE OR REPLACE VIEW scf_gold.analytics.vw_shipments AS
SELECT *
FROM scf_silver.fact.shipments;
```

```python
%sql
CREATE OR REPLACE VIEW scf_gold.analytics.vw_revenue_by_product AS
SELECT
  p.product_name,
  SUM(f.total_cost) AS total_revenue
FROM scf_silver.fact.shipments f
JOIN scf_silver.dim.product p
  ON f.product_id = p.product_id
GROUP BY p.product_name;
```

```python
%sql
CREATE OR REPLACE VIEW scf_gold.analytics.vw_revenue_by_city AS
SELECT
  origin_city,
  SUM(total_cost) AS total_revenue
FROM scf_silver.fact.shipments
GROUP BY origin_city;
```
